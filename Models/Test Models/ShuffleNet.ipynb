```python
import numpy as np
import tensorflow as tf
import keras
from keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
import opendatasets as od
from net import ShuffleNetV2

od.download("https://www.kaggle.com/datasets/kritikseth/fruit-and-vegetable-image-recognition")

training_set = tf.keras.utils.image_dataset_from_directory(
    '/content/fruit-and-vegetable-image-recognition/train',
    labels="inferred",
    label_mode="categorical",
    class_names=None,
    color_mode="rgb",
    batch_size=32,
    image_size=(224, 224),
    shuffle=True,
    seed=None,
    validation_split=None,
    subset=None,
    interpolation="bilinear",
    follow_links=False,
    crop_to_aspect_ratio=False
)

test_set = tf.keras.utils.image_dataset_from_directory(
    '/content/fruit-and-vegetable-image-recognition/validation',
    labels="inferred",
    label_mode="categorical",
    class_names=None,
    color_mode="rgb",
    batch_size=32,
    image_size=(224, 224), # adjust image size to fit ShuffleNetV2 input shape
    shuffle=True,
    seed=None,
    validation_split=None,
    subset=None,
    interpolation="bilinear",
    follow_links=False,
    crop_to_aspect_ratio=False
)

shuffle_net = tf.keras.applications.ShuffleNetV2(
    input_shape=(224, 224, 3),
    include_top=False,
    pooling='avg',
    weights=None
)

model = tf.keras.models.Sequential([
    shuffle_net,
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(units=512, activation='relu'),
    tf.keras.layers.Dense(units=256, activation='relu'),
    tf.keras.layers.Dense(units=36, activation='softmax')
])

model.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])

training_history = model.fit(x=training_set, validation_data=test_set, epochs=50)

training_loss, training_accuracy = model.evaluate(training_set)
print('Training Accuracy:', training_accuracy)

training_history = model.fit(x=training_set, validation_data=test_set, epochs=30)

training_loss, training_accuracy = model.evaluate(training_set)
print('Training Accuracy:', training_accuracy)

model.save("trained_model.h5")
```
